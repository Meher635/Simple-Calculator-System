```python
first_number=float(input("enter the first number : "))
second_number=float(input("enter the second number :"))
operators= input("enter the operator (+,-,//,%,*)")

if operators=="+":
    print(first_number+second_number)
elif operators=="-":
    print(first_number-second_number)
elif operators=="//":
    print(first_number//second_number)
elif operators=="%":
    print(first_number%second_number)
else:
    print("invalid op")
```
